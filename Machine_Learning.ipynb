Array Manipulation

```python
array = np.random.randint(0,10,(2,3))
print(array)
print(array.shape)
```

```python
#transpose of array
trans = np.transpose(array)
print(trans)
print(trans.shape)
```

```python
#alternative
trans2=array.T
print(trans2)
print(trans2.shape)
```

```python
#reshaping an array
a = np.random.randint(0,10,(2,3))
print(a)
print(a.shape)
```

```python
b = a.reshape(3,2)
print(b)
print(b.shape)
```
